```python
from mpmath import mp
from scipy.stats import chi2

def вычислить_хи_квадрат(наблюдаемые, ожидаемые):
    """Функция для вычисления значения хи-квадрат."""
    хи_квадрат = sum((наблюдаемые[i] - ожидаемые[i]) ** 2 / ожидаемые[i] for i in range(10))
    return хи_квадрат

def найти_критическое_альфа(хи_квадрат, степени_свободы):
    """Функция для нахождения критического значения альфа."""
    t = 0
    альфа1 = 0.0
    for i in range(1, 101):
        альфа = i / 100
        хи_квадрат_критическое = chi2.ppf(альфа, степени_свободы)
        if хи_квадрат < хи_квадрат_критическое and t == 0:
            альфа1 = 1 - альфа
            t = 1
        print(f"Критическое значение: {хи_квадрат_критическое:.3f}, Альфа: {1 - альфа:.2f}")
    return альфа1

def главная_функция():
    """Основная функция программы."""
    mp.dps = 100000  # количество знаков пи
    pi = mp.pi  # число пи

    n = 100000  # число обрабатываемых знаков после запятой
    наблюдаемые_частоты = [0] * 10

    # Подсчет наблюдаемых частот каждой цифры в числе пи
    for _ in range(n):
        pi *= 10
        цифра = int(pi) % 10
        наблюдаемые_частоты[цифра] += 1

    # В массиве наблюдаемые_частоты - эмпирические частоты (из числа пи)
    ожидаемые_частоты = [n // 10] * 10
    степени_свободы = 9

    # Вычисление значения хи-квадрат и вывод результата
    хи_квадрат_значение = вычислить_хи_квадрат(наблюдаемые_частоты, ожидаемые_частоты)
    print(f"Значение Хи-квадрат: {хи_квадрат_значение:.3f}")

    # Нахождение критического значения альфа и вывод результата
    критическая_альфа = найти_критическое_альфа(хи_квадрат_значение, степени_свободы)
    print(f"Критическая Альфа: {критическая_альфа:.2f}")

if __name__ == "__main__":
    главная_функция()
```
